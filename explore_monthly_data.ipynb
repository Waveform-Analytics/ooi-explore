# Explore monthly datasets

```python
import xarray as xr
import matplotlib.pyplot as plt
```

```python
# Load an example netcdf file
file_name = "data/RS03AXBS-LJ03A-09-HYDBBA302-YMD.2024.01.03.nc"
ds = xr.open_dataarray(file_name)

plt.figure(figsize=(6, 4))
ds.plot(
    cmap='viridis', 
    x='time', 
    y='frequency', 
    vmin=3, 
    vmax=6, 
    ylim=[0, 125],
    add_colorbar=False,
)
plt.title("Frequency content overview - one day at Axial Seamount");
plt.ylabel("Frequency (Hz)")
plt.xlabel("Time");
```
